```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
colspecs = [(0, 16), (18, 30), (32,38),(39,45), (46,51), (52,53), (54,60), (61,62)]
data=pd.read_fwf('psjacc463t3_mrt.txt', skiprows=26, colspecs=colspecs, names=['objID', 'mjd', 'phase', 'mag','e_mag', 'filter','reduced_mag','flags'  ])
```

```python
data
```

```python
data['objID'].unique()
```

```python
vals=((data['objID'] == '60558 Echeclus'))
obj=data[vals]
vals=((obj['filter'] == 'o'))
obj=obj[vals]
```

```python
plt.title('60558 Echeclus')
plt.errorbar(obj['mjd'], obj['mag'], yerr=obj['e_mag'], fmt='o', color='purple')
plt.xlabel('MJD(days)')
plt.ylabel('Apparemnt magnitude')
plt.gca().invert_yaxis()
```

```python
plt.title('60558 Echeclus')
plt.errorbar(obj['mjd'], obj['reduced_mag'], yerr=obj['e_mag'], fmt='o', color='cyan')
plt.xlabel('MJD(days)')
plt.ylabel('Apparent reduced magnitude')
plt.gca().invert_yaxis()
```

```python
plt.title('60558 Echeclus')
plt.errorbar(obj['phase'], obj['reduced_mag'], yerr=obj['e_mag'], fmt='o', color='black')
plt.xlabel('Phase Angle (deg)')
plt.ylabel('Apparent reduced magnitude')
plt.gca().invert_yaxis()
```
