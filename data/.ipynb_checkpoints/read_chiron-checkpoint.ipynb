```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('Chiron_Data_ATLAS_MR_Table.csv')
data['reduced_mag']=data["m"] - 5*np.log10(data["delta"]*data["r"])
```

```python
o_obs=data.loc[data['F'] =='o']
```

Let's read and plot ATLAS o-band data for Chiron 

```python
plt.title('2060 Chiron')
plt.errorbar(o_obs['MJD'], o_obs['m'], yerr=o_obs['dm'], fmt='o', color='purple')
plt.xlabel('MJD(days)')
plt.ylabel('Apparemnt magnitude')
plt.gca().invert_yaxis()
```

```python
plt.title('2060 Chiron')
plt.errorbar(o_obs['MJD'], o_obs['reduced_mag'], yerr=o_obs['dm'], fmt='o', color='cyan')
plt.xlabel('MJD(days)')
plt.ylabel('Apparent reduced magnitude')
plt.gca().invert_yaxis()
```

```python
plt.title('2060 Chiron')
plt.errorbar(o_obs['alpha'], o_obs['reduced_mag'], yerr=o_obs['dm'], fmt='o', color='black')
plt.xlabel('Phase(deg)')
plt.ylabel('Apparent reduced magnitude')
plt.gca().invert_yaxis()
w=(o_obs['MJD']>=59383)
plt.errorbar(o_obs[w]['alpha'], o_obs[w]['reduced_mag'], yerr=o_obs[w]['dm'], fmt='o', color='red')
```
